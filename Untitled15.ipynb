```python
def fahrenheit_to_kelvin(fahrenheit):
  return (fahrenheit - 32) * 5/9 + 273.15

fahrenheit_temperatures = [212, 32, 68, -4, 100]


kelvin_temperatures = [fahrenheit_to_kelvin(f) for f in fahrenheit_temperatures]

print("Fahrenheit в Kelvin :")
print("-" * 30)
print("Fahrenheit | Kelvin")
print("-" * 30)

for fahrenheit, kelvin in zip(fahrenheit_temperatures, kelvin_temperatures):
  print(f"{fahrenheit:10.2f} | {kelvin:6.2f}")

print("-" * 30)

max_kelvin = max(kelvin_temperatures)
min_kelvin = min(kelvin_temperatures)
print(f"\nMaximum Kelvin temperature: {max_kelvin:.2f}")
print(f"Minimum Kelvin temperature: {min_kelvin:.2f}")

avg_kelvin = sum(kelvin_temperatures) / len(kelvin_temperatures)
print(f"Average Kelvin temperature: {avg_kelvin:.2f}")

temperature_range = max_kelvin - min_kelvin
print(f"Temperature range (Kelvin): {temperature_range:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 100 * np.random.rand(50)
colors = np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.colorbar(label='Color Value')
plt.title('Диаграмма рассеяния с различными размерами и цветами')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()


x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap='magma')
plt.colorbar(label='Z-value')
plt.title('заполненный контурный график')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()


x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.stem(x, y, linefmt='b-', markerfmt='ro', basefmt='k--')
plt.title('Стволовый график')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt # Создаем данные для цветка
theta = np.linspace(0, 2 * np.pi, 100)
r = 1 + 0.5 * np.sin(5 * theta) # Полярные координаты
x = r * np.cos(theta)
y = r * np.sin(theta) # Рисуем цветок
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='magenta')
plt.fill(x, y, color='pink', alpha=0.5)
plt.title('Цветок с помощью Matplotlib')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers # Загрузка набора данных
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data() # Названия классов

class_names = [ "Топ", "Брюки", "Свитер", "Платье", "Пальто", "Сандалии", "Рубашка", "Кроссовки", "Сумка", "Ботинки" ] # Нормируем данные
train_images = train_images / 255.0
test_images = test_images / 255.0 # Строим модель
model = keras.Sequential([ layers.Flatten(input_shape=(28, 28)), layers.Dense(128, activation='relu'), layers.Dense(10, activation='softmax') ]) # Компилируем модель
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # Обучаем модель
model.fit(train_images, train_labels, epochs=5) # Оцениваем модель
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f'Accuracy: {test_acc}') # Показать несколько примеров
plt.figure(figsize=(10, 10))
for i in range(9):
  plt.subplot(3, 3, i + 1)
  plt.imshow(test_images[i], cmap='gray')
  plt.title(class_names[test_labels[i]])
  plt.axis('off')
plt.show()
```
